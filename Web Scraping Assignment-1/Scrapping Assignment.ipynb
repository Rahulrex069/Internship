# 1. Write a python program to display all the header tags from ‘en.wikipedia.org/wiki/Main_Page’.

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
page = requests.get('https://en.wikipedia.org/wiki/Main_Page')
page
```

```python
soup1 = BeautifulSoup(page.content)
soup1
```

```python
h1_tags = soup1.find_all('h1',class_='firstHeading')
h1_tags
```

```python
h1_head_tags = []

for i in h1_tags:
    h1_head_tags.append(i.text.replace("/n",''))
h1_head_tags
```

```python
h2_tags = soup1.find_all('h2', class_='mp-h2')
h2_tags
```

```python
h2_head_tags = []

for i in h2_tags:
    h2_head_tags.append(i.text.replace("/n",''))
h2_head_tags
```

```python
h3_tags =soup1.find_all('h3', class_='vector-menu-heading')
h3_tags
```

```python
h3_head_tags = []

for i in h3_tags:
    h3_head_tags.append(i.text.strip())
h3_head_tags
```

```python
print(len(h1_head_tags),len(h2_head_tags),len(h3_head_tags))
```

```python
Header_tags=pd.DataFrame({'h3_head_tags': h3_head_tags})
Header_tags['h2_head_tags']=pd.Series(h2_head_tags)
Header_tags['h1_head_tags']=pd.Series(h1_head_tags)
Header_tags
```
