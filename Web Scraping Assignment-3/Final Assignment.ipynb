```python
URL1 = driver.find_elements_by_xpath('//ul[@class="a-pagination"]/li[2]/a')
URL2 = driver.find_elements_by_xpath('//ul[@class="a-pagination"]/li[3]/a')
URL3 = driver.find_elements_by_xpath('//ul[@class="a-pagination"]/li[4]/a')
```

```python
URL_List = []

for i in URL1:
    URL_List.append(i.get_attribute('href'))

for i in URL2:
    URL_List.append(i.get_attribute('href'))

for i in URL3:
    URL_List.append(i.get_attribute('href'))
    
URL_List
```

```python
product_URL = []

for i in URL_List:
    driver.get(i)
    url=driver.find_elements_by_xpath("//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-2']/a")
    for i in url: 
        product_URL.append(i.get_attribute('href'))
    
product_URL
```

```python
print(len(product_URL))
```

```python
Brand = []
Price = []
Rating = []
No_of_Ratings = []
Return_or_Exchange = []
Expected_Delivery = []
Availability = []


for i in product_URL:
    driver.get(i)
    try:   # Brand
        p=driver.find_element_by_id('productTitle')
        Brand.append(p.text)
    except NoSuchElementException:
        Brand.append("-")
    
    try: #Price
        P=driver.find_element_by_xpath('//td[@class="a-span12"]/span')
        Price.append(P.text)
    except NoSuchElementException:
        Price.append("-")
        

    try: # Rating
        R=driver.find_element_by_xpath("//span[@class='a-size-medium a-color-base']")
        Rating.append(R.text)
    except NoSuchElementException:
        Rating.append("-")
        
    try:
        Rat=driver.find_element_by_xpath("//div[@data-hook='total-review-count']/span")
        No_of_Ratings.append(Rat.text)
    except NoSuchElementException:
        No_of_Ratings.append("-")
        
    try:
        Ret=driver.find_element_by_xpath("//div[@data-name = 'RETURNS_POLICY']/span/div[2]/a")
        Return_or_Exchange.append(Ret.text)
    except NoSuchElementException:
        Return_or_Exchange.append("-")
        
    try:
        ED=driver.find_element_by_xpath("//div[@id = 'ddmDeliveryMessage']/b")
        Expected_Delivery.append(ED.text)
    except NoSuchElementException:
        Expected_Delivery.append("-")
        
    try:
        A=driver.find_element_by_xpath("//div[@id = 'availability']/span")
        Availability.append(A.text)
    except NoSuchElementException:
        Availability.append("-")
        
```

```python
Amazon=pd.DataFrame({})
Amazon['Brand & Product']=Brand
Amazon['Price']=Price
Amazon['Rating']=Rating
Amazon['No.Of Ratings']=No_of_Ratings
Amazon['Return/Exchange']=Return_or_Exchange
Amazon['Expected Delivery']=Expected_Delivery
Amazon['Availability']=Availability
Amazon.index = np.arange(1, len(Amazon)+1)
Amazon
```
