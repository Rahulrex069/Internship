# Q5. Scrape the details of trending repositories on Github.com.
Url = https://github.com/
You have to find the following details:
A) Repository title
B) Repository description
C) Contributors count
D) Language used
Note: - From the home page you have to click on the trending option from Explore menu through code.

```python
driver = webdriver.Chrome(executable_path='/Users/rahulmore/chromedriver')
driver
```

```python
URL = driver.get('https://github.com/')
URL
driver.maximize_window()
time.sleep(5)
```

```python
button = driver.find_element_by_xpath("//ul[@class = 'd-lg-flex list-style-none']/li[4]/details/summary")
button.click()
time.sleep(2)
```

```python
button2= driver.find_element_by_xpath("//ul[@class = 'list-style-none mb-3'][2]/li[3]/a")
button2.click()
time.sleep(2)
```

```python
URLs = driver.find_elements_by_xpath("//h1[@class = 'h3 lh-condensed']/a")

URL_List = []

for i in URLs:
    URL_List.append(i.get_attribute('href'))
    
URL_List
```

```python
Repository_Name = []

for i in URL_List:
    driver.get(i)
    url=driver.find_elements_by_xpath("//strong[@class='mr-2 flex-self-stretch']/a")
    for i in url: 
        Repository_Name.append(i.text)
        
Repository_Name
```

```python
Repository_Description = []



for i in URL_List:
    driver.get(i)
    try:
        P=driver.find_element_by_xpath("//div[@class='BorderGrid-cell']/p")
        Repository_Description.append(P.text)
    except NoSuchElementException:
        Repository_Description.append("Description Not available")
        
Repository_Description
```

```python
Contributors = []

for i in URL_List:
    driver.get(i)
    try:
        P=driver.find_element_by_xpath("//h2[@class='h4 mb-3']/a[contains(text(),'Contributors')]/span")
        Contributors.append(P.text)
    except:
        Contributors.append('NA')
Contributors
```

```python
all_data = []
for i in URL_List:
    
    driver.get(i)
    languages_used_elemens = driver.find_elements_by_xpath('//a[contains(@data-ga-click,"Repository, language")]')
    
    language_used = []
    for lang in languages_used_elemens:
        language_used.append(lang.text.split('\n')[0])
    all_data.append(language_used,)
    
all_data
```

```python
github=pd.DataFrame({})
github['Repository Name']=Repository_Name
github['Repository_Description']=Repository_Description
github['Contributors count']=Contributors
github['Language used']=all_data

github.index = np.arange(1, len(github)+1)
github
```

```python
driver.close()
```

```python
Name = []
Skills = []
Designation = []
Company = []
URLs = []

while(True):
    P=driver.find_elements_by_xpath("//span[@class='fl ellipsis']")
    try: #Name 
        for i in P:
            Name.append(i.text)
    except NoSuchElementException:
        Name.append('No Data')
        
        
    S=driver.find_elements_by_xpath("//div[@class = 'hireSec highlightable']")
    try:
        for i in S:
            Skills.append(i.text)
    except NoSuchElementException:
        Skills.append('No Data')
        
        
    D=driver.find_elements_by_xpath("//span[@class = 'ellipsis clr']")
    try:
        for i in D:
            Designation.append(i.text)
    except NoSuchElementException:
        Designation.append('No Data')
        
        
    C=driver.find_elements_by_xpath("//div[@class = 'vcard']/p/a[2]")
    try:
        for i in C:
            Company.append(i.text)
    except NoSuchElementException:
        Company.append('No Data')
        
        
    U=driver.find_elements_by_xpath("//p[@class = 'highlightable']/a[1]")
    for i in U:
        URLs.append(i.get_attribute('href'))
        
    try:
        nex = driver.find_element_by_xpath("//button[@class = 'grayBtn'][contains(text(),'Next')]").click()
    except:
        break
```

```python
Location = []

for i in URLs:
    driver.get(i)
    try:
        L = driver.find_element_by_xpath('//div[@class = "rFrame fl infoWrapper"]/div[5]/a')
        if L.text is None:
            Location.append('-')
        else:
            Location.append(L.text)
    except NoSuchElementException:
        Location.append('No Data')
        
Location
```

```python
Naukri=pd.DataFrame({})
Naukri['Name']=Name
Naukri['Designation']=Designation
Naukri['Company']=Company
Naukri['Skills they hire for']=Skills
Naukri['Location']=Location

Naukri.index = np.arange(1, len(Naukri)+1)
Naukri
```
